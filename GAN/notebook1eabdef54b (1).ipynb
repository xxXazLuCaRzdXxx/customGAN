```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import time

# Define Hyperparameters
EPOCHS = 100 # 
LATENT_DIM = 100
BATCH_SIZE = 128 
BUFFER_SIZE = 60000 

IMAGE_WIDTH = 28
IMAGE_HEIGHT = 28
IMAGE_CHANNELS = 1
IMAGE_SHAPE = (IMAGE_HEIGHT, IMAGE_WIDTH, IMAGE_CHANNELS)


SAVE_EVERY_N_EPOCHS = 10 

# 1. Load and Prepare the MNIST Dataset
(train_images, _), (_, _) = tf.keras.datasets.mnist.load_data()

# Reshape and Normalize images to [-1, 1]
train_images = train_images.reshape(train_images.shape[0], IMAGE_HEIGHT, IMAGE_WIDTH, IMAGE_CHANNELS).astype('float32')
train_images = (train_images - 127.5) / 127.5  

# Batch and shuffle the data
train_dataset = tf.data.Dataset.from_tensor_slices(train_images).shuffle(BUFFER_SIZE).batch(BATCH_SIZE)

# 2. Build the Generator Model
def make_generator_model():
    model = tf.keras.Sequential()
    # Start with a Dense layer that projects the latent vector to a suitable starting shape for conv layers
    model.add(tf.keras.layers.Dense(7*7*256, use_bias=False, input_shape=(LATENT_DIM,)))
    model.add(tf.keras.layers.BatchNormalization())
    model.add(tf.keras.layers.LeakyReLU())

    model.add(tf.keras.layers.Reshape((7, 7, 256)))
    assert model.output_shape == (None, 7, 7, 256)  

    # Upsample to 14x14
    model.add(tf.keras.layers.Conv2DTranspose(128, (5, 5), strides=(1, 1), padding='same', use_bias=False))
    assert model.output_shape == (None, 7, 7, 128)
    model.add(tf.keras.layers.BatchNormalization())
    model.add(tf.keras.layers.LeakyReLU())

    # Upsample to 28x28
    model.add(tf.keras.layers.Conv2DTranspose(64, (5, 5), strides=(2, 2), padding='same', use_bias=False))
    assert model.output_shape == (None, 14, 14, 64)
    model.add(tf.keras.layers.BatchNormalization())
    model.add(tf.keras.layers.LeakyReLU())

    # Final upsample to image size (28x28x1)
    model.add(tf.keras.layers.Conv2DTranspose(IMAGE_CHANNELS, (5, 5), strides=(2, 2), padding='same', use_bias=False, activation='tanh'))
    assert model.output_shape == (None, IMAGE_HEIGHT, IMAGE_WIDTH, IMAGE_CHANNELS)

    return model

generator = make_generator_model()
generator.summary()

# 3. Build the Discriminator Model
def make_discriminator_model():
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Conv2D(64, (5, 5), strides=(2, 2), padding='same',
                                     input_shape=IMAGE_SHAPE)) # (28,28,1) -> (14,14,64)
    model.add(tf.keras.layers.LeakyReLU())
    model.add(tf.keras.layers.Dropout(0.3))

    model.add(tf.keras.layers.Conv2D(128, (5, 5), strides=(2, 2), padding='same')) # (14,14,64) -> (7,7,128)
    model.add(tf.keras.layers.LeakyReLU())
    model.add(tf.keras.layers.Dropout(0.3))

    model.add(tf.keras.layers.Flatten())
    model.add(tf.keras.layers.Dense(1)) # No sigmoid here, using from_logits=True in loss

    return model

discriminator = make_discriminator_model()
discriminator.summary() 


# 4. Define Loss Functions and Optimizers

# This method returns a helper function to compute cross entropy loss
cross_entropy = tf.keras.losses.BinaryCrossentropy(from_logits=True)

def discriminator_loss(real_output, fake_output):
    real_loss = cross_entropy(tf.ones_like(real_output), real_output)
    fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
    total_loss = real_loss + fake_loss
    return total_loss

def generator_loss(fake_output):
    # Generator wants the discriminator to think the fake images are real (label 1)
    return cross_entropy(tf.ones_like(fake_output), fake_output)

generator_optimizer = tf.keras.optimizers.Adam(1e-4)
discriminator_optimizer = tf.keras.optimizers.Adam(1e-4)


# 5. Define the Training Step
@tf.function
def train_step(images):
    noise = tf.random.normal([BATCH_SIZE, LATENT_DIM])

    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
        generated_images = generator(noise, training=True)

        real_output = discriminator(images, training=True)
        fake_output = discriminator(generated_images, training=True)

        gen_loss = generator_loss(fake_output)
        disc_loss = discriminator_loss(real_output, fake_output)

    gradients_of_generator = gen_tape.gradient(gen_loss, generator.trainable_variables)
    gradients_of_discriminator = disc_tape.gradient(disc_loss, discriminator.trainable_variables)

    generator_optimizer.apply_gradients(zip(gradients_of_generator, generator.trainable_variables))
    discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, discriminator.trainable_variables))
    
    return gen_loss, disc_loss

# Helper function to generate and save images
def generate_and_save_images(model, epoch, test_input):
    # `training` is set to False.
    # This is so all layers run in inference mode (e.g. BatchNormalization).
    predictions = model(test_input, training=False)

    fig = plt.figure(figsize=(6, 6))

    for i in range(predictions.shape[0]): 
        plt.subplot(4, 4, i + 1)
        # De-normalize the image from [-1, 1] to [0, 1] for displaying
        plt.imshow((predictions[i, :, :, 0] * 127.5 + 127.5) / 255.0, cmap='gray')
        plt.axis('off')

    plt.suptitle(f"Generated Images at Epoch: {epoch:04d}", fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95]) 
    plt.show()
    plt.close(fig) 

num_examples_to_generate = 16
seed = tf.random.normal([num_examples_to_generate, LATENT_DIM])


# 6. Training Loop
def train(dataset, epochs):
    gen_losses = []
    disc_losses = []

    for epoch in range(epochs):
        start = time.time()
        epoch_gen_loss_avg = tf.keras.metrics.Mean()
        epoch_disc_loss_avg = tf.keras.metrics.Mean()

        for image_batch in dataset:
            g_loss, d_loss = train_step(image_batch)
            epoch_gen_loss_avg.update_state(g_loss)
            epoch_disc_loss_avg.update_state(d_loss)
        
        gen_losses.append(epoch_gen_loss_avg.result())
        disc_losses.append(epoch_disc_loss_avg.result())

        print(f'Time for epoch {epoch + 1} is {time.time()-start:.2f} sec. Gen Loss: {epoch_gen_loss_avg.result():.4f}, Disc Loss: {epoch_disc_loss_avg.result():.4f}')

        # Produce images for the GIF as we go
        if (epoch + 1) % SAVE_EVERY_N_EPOCHS == 0 or epoch == epochs -1:
            generate_and_save_images(generator, epoch + 1, seed)
            
    # Generate after the final epoch
    generate_and_save_images(generator, epochs, seed)
    
    return gen_losses, disc_losses

print("Starting Training...")
gen_losses, disc_losses = train(train_dataset, EPOCHS)
print("Training Finished.")

# Plot training losses
plt.figure(figsize=(10, 5))
plt.plot(gen_losses, label='Generator Loss')
plt.plot(disc_losses, label='Discriminator Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('GAN Training Losses')
plt.show()

# 7. Generate Novel Images from the Trained Model
print("\n--- Generating Novel Images from the Trained Model ---")

def generate_novel_images(trained_generator, num_images=16):
    # Generate new random noise vectors
    random_latent_vectors = tf.random.normal(shape=[num_images, LATENT_DIM])
    
    # Generate images
    predictions = trained_generator(random_latent_vectors, training=False)
    
    # Plot the images
    fig = plt.figure(figsize=(8, 8))
    plt.suptitle("Novel Generated Images", fontsize=16)
    for i in range(predictions.shape[0]):
        plt.subplot(4, 4, i + 1) 
        img_to_show = (predictions[i, :, :, 0] * 127.5 + 127.5) / 255.0
        plt.imshow(img_to_show, cmap='gray')
        plt.axis('off')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

# Use the trained generator global variable
generate_novel_images(generator, num_images=16)

print("\n--- Generating a specific number (e.g., trying to find a '7') ---")
print("This is more advanced and often involves searching the latent space or conditional GANs.")
print("With an unconditional GAN like this, you generate randomly and see what you get.")
print("To get a specific digit, you would typically train a Conditional GAN (CGAN).")
print("For now, we can just generate a larger batch and visually inspect.")

generate_novel_images(generator, num_images=25) 
```

```python
print("\n--- Demonstrating Latent Space Interpolation ---")

def interpolate_latent_space(trained_generator, z1, z2, num_steps=10):
    """
    Generates images by interpolating between two latent vectors z1 and z2.

    Args:
        trained_generator: The trained GAN generator model.
        z1: The first latent vector (shape [1, LATENT_DIM]).
        z2: The second latent vector (shape [1, LATENT_DIM]).
        num_steps: The number of intermediate images to generate between z1 and z2.

    Returns:
        A list of generated images.
    """
    alphas = np.linspace(0, 1, num_steps) # Create num_steps points from 0 to 1
    interpolated_vectors = []
    for alpha in alphas:
        z_interp = alpha * z2 + (1 - alpha) * z1 # Linear interpolation
        interpolated_vectors.append(z_interp)
    
    # Stack the list of vectors into a single batch tensor
    interpolated_vectors_batch = tf.concat(interpolated_vectors, axis=0)
    
    # Generate images from the interpolated latent vectors
    # Ensure the generator is called with training=False
    interpolated_images = trained_generator(interpolated_vectors_batch, training=False)
    
    return interpolated_images

# 1. Generate two random latent vectors to interpolate between
z_start = tf.random.normal([1, LATENT_DIM])
z_end = tf.random.normal([1, LATENT_DIM])

# 2. Generate images at the start and end points to see what they look like
start_image = generator(z_start, training=False)
end_image = generator(z_end, training=False)

# 3. Perform the interpolation
num_interpolation_steps = 10 
interpolated_images_sequence = interpolate_latent_space(generator, z_start, z_end, num_steps=num_interpolation_steps)

# 4. Visualize the interpolation
print(f"\nVisualizing interpolation between two latent vectors over {num_interpolation_steps} steps:")

fig_height = 4 
fig_width = num_interpolation_steps * 2 
fig = plt.figure(figsize=(fig_width, fig_height))
plt.suptitle("Latent Space Interpolation", fontsize=16)

for i in range(num_interpolation_steps):
    plt.subplot(1, num_interpolation_steps, i + 1)
    # De-normalize from [-1,1] to [0,1] for imshow
    img_to_show = (interpolated_images_sequence[i, :, :, 0] * 127.5 + 127.5) / 255.0
    plt.imshow(img_to_show, cmap='gray')
    plt.axis('off')
    if i == 0:
        plt.title("Start (z1)")
    elif i == num_interpolation_steps - 1:
        plt.title("End (z2)")
    else:
        plt.title(f"Step {i}")

plt.tight_layout(rect=[0, 0.03, 1, 0.93]) 
plt.show()


print("\n--- Another Interpolation Example ---")
z_start_2 = tf.random.normal([1, LATENT_DIM])
z_end_2 = tf.random.normal([1, LATENT_DIM])
interpolated_images_sequence_2 = interpolate_latent_space(generator, z_start_2, z_end_2, num_steps=12)

fig = plt.figure(figsize=(12*2, 4))
plt.suptitle("Latent Space Interpolation - Example 2", fontsize=16)
for i in range(12):
    plt.subplot(1, 12, i + 1)
    img_to_show = (interpolated_images_sequence_2[i, :, :, 0] * 127.5 + 127.5) / 255.0
    plt.imshow(img_to_show, cmap='gray')
    plt.axis('off')
plt.tight_layout(rect=[0, 0.03, 1, 0.93])
plt.show()
```
